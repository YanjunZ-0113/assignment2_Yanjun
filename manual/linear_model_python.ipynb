This notebook demonstrates a simple linear regression analysis using python to model Salary based on Years of Experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv("regression_data.csv")
dataset
```

This is used to read the table

```python
x = dataset.iloc[:,0]
y = dataset.iloc[:,1]
plt.scatter(x, y, color="blue")
plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.show()
```

This is used to create a scatter plot

```python
model = LinearRegression()
model.fit(dataset.iloc[:, [0]], dataset.iloc[:,[1]])
plt.scatter(x, y, color="blue")
plt.plot(x, model.predict(dataset.iloc[:, [0]]), color="red")
plt.title("Salary vs Experience")
plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.show()
```

This code creat and train a linear regression modelfits and then plots a linear regression model

```python
r = model.score(dataset.iloc[:, [0]], dataset.iloc[:,[1]])
print("R-squared is",r)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
```

### This assignment adds Pearson’s r and MSE to evaluate the regression model and includes these values in the plot.

```python
dataset = pd.read_csv("regression_data.csv")
dataset
```

### load data from CSV data.file

```python
x = dataset.iloc[:, 0]
y = dataset.iloc[:, 1]
```

This is used to define X and Y. 

```python
slope, intercept, r_value, p_value, std_err = linregress(x, y)

y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)

print("Slope:", slope)
print("Intercept:", intercept)
print("Correlation coefficient:", r_value)
print("Mean Squared Error:", mse)

plt.scatter(x, y, color="blue")
plt.plot(x, y_pred, color="red")
plt.text(1.5, max(y) - 14000,
         f"y = {slope:.2f}x + {intercept:.2f}\n"
         f"r = {r_value:.2f}\nMSE = {mse:.2f}",
         fontsize=12)

plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.title("Regression")
plt.show()
```
