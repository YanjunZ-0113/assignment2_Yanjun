This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.

```r
dataset <- read.csv("regression_data.csv")
```

read dataset

```r
plot(dataset$YearsExperience, dataset$Salary, col="red")
```

Make a scotter plot

```r
model <- lm(Salary ~ YearsExperience, data=dataset)
```

Fit a linear regression model

```r
library(ggplot2)
ggplot() +
  geom_point(aes(x = dataset$YearsExperience, y = dataset$Salary), colour = 'red') +
  geom_line(aes(x = dataset$YearsExperience, y = predict(model, newdata = dataset)), colour = 'blue') +
  ggtitle('Salary vs Experience') +
  xlab('Years of experience') +
  ylab('Salary')
```

Plot the data points and fitted regression line

```r
summary(model)
```

```r
library(ggplot2)

dataset <- read.csv("regression_data.csv")
```

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
slope <- coef(model)[2]
intercept <- coef(model)[1]
pred <- predict(model)
r <- cor(dataset$YearsExperience, dataset$Salary)
mse <- mean((dataset$Salary - pred)^2)

cat("Slope:", slope)
cat("Intercept:", intercept)
cat("Correlation coefficient:", r)
cat("Mean Squared Error:", mse)
```

This used to get the each value use the model. 

```r
ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
  geom_point(colour = "red") +
  geom_smooth(method = "lm", formula = y ~ x, se = FALSE, colour = "blue") +
  annotate(
    "text",
    x = 1.5,
    y = max(dataset$Salary) - 10000,
    label = paste(
      "y =", round(slope, 2), "x +", round(intercept, 2),
      "\nr =", round(r, 2),
      "\nMSE =", round(mse, 2)
    ),
    size = 4
  ) +
  ggtitle("Salary vs Experience") +
  xlab("Years of Experience") +
  ylab("Salary")
```
