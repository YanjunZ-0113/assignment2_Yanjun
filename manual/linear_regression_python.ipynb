This notebook demonstrates a simple linear regression analysis using python to model Salary based on Years of Experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv("regression_data.csv")
dataset
```

This is used to read the table

```python
x = dataset.iloc[:,0]
y = dataset.iloc[:,1]
plt.scatter(x, y, color="blue")
plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.show()
```

This is used to create a scatter plot

```python
model = LinearRegression()
model.fit(dataset.iloc[:, [0]], dataset.iloc[:,[1]])
plt.scatter(x, y, color="blue")
plt.plot(x, model.predict(dataset.iloc[:, [0]]), color="red")
plt.title("Salary vs Experience")
plt.xlabel(dataset.columns[0])
plt.ylabel(dataset.columns[1])
plt.show()
```

This code creat and train a linear regression modelfits and then plots a linear regression model

```python
r = model.score(dataset.iloc[:, [0]], dataset.iloc[:,[1]])
print("R-squared is",r)
```
