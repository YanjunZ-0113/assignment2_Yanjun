# Linear Regression in R

This notebook analyzes salary data with simple linear regression, visualizes the relationship between years of experience and salary, fits a predictive model, and summarizes model performance.

## Step 1: Load the data

Read `regression_data.csv` and display the dataset.

```r
data <- read.csv("regression_data.csv", stringsAsFactors = FALSE)
data
```

## Step 2: Visualize the relationship

Create a scatter plot of years of experience versus salary to inspect the trend visually.

```r
plot(
  data$YearsExperience,
  data$Salary,
  xlab = "Years of Experience",
  ylab = "Salary",
  main = "Salary vs Years of Experience",
  pch = 19,
  col = "steelblue"
)
```

## Step 3: Fit a linear regression model

Fit a simple linear model predicting salary from years of experience using `lm()`.

```r
model <- lm(Salary ~ YearsExperience, data = data)
coef(model)
```

## Step 4: Add the regression line

Overlay the fitted regression line on the scatter plot with clear labels and a legend.

```r
plot(
  data$YearsExperience,
  data$Salary,
  xlab = "Years of Experience",
  ylab = "Salary",
  main = "Salary vs Years of Experience with Regression Line",
  pch = 19,
  col = "steelblue"
)
abline(model, col = "red", lwd = 2)
legend(
  "topleft",
  legend = c("Observed", "Regression line"),
  col = c("steelblue", "red"),
  pch = c(19, NA),
  lty = c(NA, 1),
  lwd = c(NA, 2)
)
```

## Step 5: Evaluate and summarize the model

Review the full model summary, R-squared, RMSE, and a sample prediction.

```r
model_summary <- summary(model)
model_summary

r2 <- model_summary$r.squared
rmse <- sqrt(mean(residuals(model)^2))
predict_5yr <- predict(model, newdata = data.frame(YearsExperience = 5.0))

cat(sprintf("R-squared: %.6f\n", r2))
cat(sprintf("RMSE: %.4f\n", rmse))
cat(sprintf("Predicted salary at 5 years experience: %.2f\n", predict_5yr))
```
